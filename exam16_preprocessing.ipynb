### 데이터변환

```python
df = pd.read_csv('./datasets/auto-mpg.csv', header = None)
df.columns = ['mpg', 'cylinder', 'displacement', 'horsepower', 'weight',
      'acceleration', 'model tyer', 'origin', 'name']
df
```

### 단위환산

```python
mpg_to_kpl_rate = 0.425144
df['kpl'] = df['mpg'] * mpg_to_kpl_rate
df['kpl'] = df['kpl'].round(2)
df
```

```python
df.info()
```

```python
df['horsepower'].unique()
```

```python
df['horsepower'].value_counts().sort_index()
```

```python
df['horsepower'].replace('?', np.nan, inplace=True)
df.dropna(subset=['horsepower'], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df['horsepower'] = df['horsepower'].astype('float')
df.info()
```

```python
df['origin'].unique()
```

```python
df['origin'].replace({1:'USA', 2:'EU', 3:'JP'}, inplace=True)
df
```
