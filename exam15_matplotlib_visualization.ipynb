```python
cmaps = {}

gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))


def plot_color_gradients(category, cmap_list):
    # Create figure and adjust figure height to number of colormaps
    nrows = len(cmap_list)
    figh = 0.35 + 0.15 + (nrows + (nrows - 1) * 0.1) * 0.22
    fig, axs = plt.subplots(nrows=nrows + 1, figsize=(6.4, figh))
    fig.subplots_adjust(top=1 - 0.35 / figh, bottom=0.15 / figh,
                        left=0.2, right=0.99)
    axs[0].set_title(f'{category} colormaps', fontsize=14)

    for ax, name in zip(axs, cmap_list):
        ax.imshow(gradient, aspect='auto', cmap=mpl.colormaps[name])
        ax.text(-0.01, 0.5, name, va='center', ha='right', fontsize=10,
                transform=ax.transAxes)

    # Turn off *all* ticks & spines, not just the ones with colormaps.
    for ax in axs:
        ax.set_axis_off()

    # Save colormap list for later.
    cmaps[category] = cmap_list
```

```python
fig = plt.figure(figsize=(4, 3))
ax = fig.add_axes([0, 0, 1, 1])
p1 = ax.bar(idx, dog_means, width, yerr=dog_std, label='Dog')
p2 = ax.bar(idx, cat_means, width, yerr=cat_std, label='Cat', bottom=dog_means)
ax.bar_label(p1, label_type='center')
ax.bar_label(p2, label_type='center')
ax.bar_label(p2)
ax.axhline(color='gray', linewidth=0.8)
ax.set_ylabel('Score')
ax.set_xlabel('groups')
ax.legend()
plt.show()
```

```python
import seaborn as sns
sns.set(style='darkgrid')

fig, ax = plt.subplots(1, 2, figsize = (8, 4))
ax[0].scatter(x = 'sepal length (cm)', y = 'petal width (cm)',
              data = iris_df, color = 'gray') #scatter 산점도
ax[0].set_title('Before')

ax[1].scatter(x = 'sepal length (cm)', y = 'petal width (cm)',
              data = iris_df[iris_df['species'] == 'setosa'], 
                             color = 'olive', label = 'setosa', s = 10, alpha = 0.7)
ax[1].scatter(x = 'sepal length (cm)', y = 'petal width (cm)',
              data = iris_df[iris_df['species'] == 'versicolor'], 
                             color = 'palevioletred', label = 'versicolor', s = 10, alpha = 0.7)
ax[1].scatter(x = 'sepal length (cm)', y = 'petal width (cm)',
              data = iris_df[iris_df['species'] == 'virginica'], 
                             color = 'plum', label = 'virginica', s = 10, alpha = 0.7)
ax[1].legend()
ax[1].set_xlabel('sepal length')
ax[1].set_ylabel('petal width')
ax[1].spines['top'].set_visible(False) # 그래프의 상단 선을 없애준다
ax[1].spines['right'].set_visible(False) # 그래프의 우측 선을 없애준다
# alpha = 0.7 -> 투명도 70 -> 겹쳐있는 점들은 진하게 출력
ax[1].set_title('After')

plt.show()
```

```python
# 히스토그램

color_map = ['gray'] * 20
color_map[9] = color_map[10] = '#3caea3'

n_point = 100000
n_bins = 20

dist1 = np.random.normal(0, 2, n_point) # normal -> 정규분포 /평균 0, 편차 2
dist2 = np.random.normal(0, 10, n_point) # 평균 0, 편차 10

fig, axes = plt.subplots(1, 2, sharey = True, tight_layout = True, figsize = (7, 4))
n, bins, patches = axes[0].hist(dist1, bins = n_bins)
axes[0].set_xlim(-10, 20) #-> x축 스케일을 지정
for i, patch in enumerate(patches):
    patch.set_facecolor(color_map[i])
axes[1].hist(dist2, bins = n_bins)
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(3, 9), tight_layout=True)
axes[0].hist2d(dist1, dist2, bins=40)
axes[1].hist2d(dist1, dist2, bins=(80, 10), norm=mpl.colors.LogNorm())
plt.show()
```

```python
fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(projection='3d')

hist, xedges, yedges = np.histogram2d(dist1, dist2, bins=40, range=[[-7, 7], [-10, 20]])
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing='ij')
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0
dx = dy = 0.3 * np.ones_like(zpos)
dz = hist.ravel()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')
ax.view_init(elev=30.0, azim=30)
plt.show()
```
