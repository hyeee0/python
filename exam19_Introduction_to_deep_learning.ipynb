## linear regression

### 기존의 프로그램 방식

```python
def celsius_to_fahenheit(x):
    return x * 1.8 + 32
```

```python
celsius_value = int(input('화씨온도로 변환할 섭씨온도를 입력하세요.\n'))
print('화씨온도로', celsius_to_fahenheit(20))
```

### 머신러닝 방식

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_C = np.array(range(0, 100))
data_F = celsius_to_fahenheit(data_C)
print(data_C)
print(data_F)
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(1,)))
model.add(Dense(1))
model.compile(loss='mse', optimizer='rmsprop')
model.summary()
```

```python
scaled_data_C = data_C / 100
scaled_data_F = data_F / 100
print(scaled_data_C)
print(scaled_data_F)
```

```python
print(model.predict([0.01]))
```

```python
model.save('before_learning.h5')
```

```python
fit_hist = model.fit(scaled_data_C, scaled_data_F, epochs=1000)
```

```python
model.predict([0.01])
```

```python
model.save('after_learning.h5')
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
model.predict([0.2]) * 100
```

## 잡음이 있는 현실 데이터

```python
noise = np.random.normal(0, 0.05, 100)
noise
```

```python
noised_scaled_data_F = np.array([])
for data in scaled_data_F:
    noised_scaled_data_F = np.append(noised_scaled_data_F, np.random.normal(0, 0.05, 100) + data)
print(noised_scaled_data_F)
print(len(noised_scaled_data_F))
```

```python
noised_scaled_data_C = []
for data in scaled_data_C:
    for _ in range(100):
        noised_scaled_data_C.append(data)
noised_scaled_data_C = np.array(noised_scaled_data_C)
print(noised_scaled_data_C)
print(len(noised_scaled_data_C))
```

```python
plt.scatter(x=noised_scaled_data_C, y=noised_scaled_data_F)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(50, 50))
ax = fig.add_subplot()
ax.scatter(x=noised_scaled_data_C, y=noised_scaled_data_F, alpha=0.2, s=200, marker='+')
plt.show()
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(1,)))
model.add(Dense(1))
model.compile(loss='mse', optimizer='rmsprop')
model.summary()
```

```python
model.predict([0.01])
```

```python
fit_hist = model.fit(noised_scaled_data_C, noised_scaled_data_F, epochs=20)
```

```python
model.save('noised_after_learning.h5')
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
model.predict([0.01])
```
