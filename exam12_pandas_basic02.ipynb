```python
import pandas as pd
```

```python
exam_data = {'이름':['경석', '연주', '선희'], 
             '수학':[90, 85, 70],
             '영어':[98, 88, 78],
             '음악':[85, 79, 75],
             '체육':[70, 91, 84]}

df = pd.DataFrame(exam_data)
df
```

```python
df.set_index('이름', inplace = True) #이름이 인덱스로 들어가있다.
df
```

```python
df.shape #shape 은 데이터의 모양 -> 행은 3 열은 4
```

```python
df.iloc[0,3] = 90 # 1행4열의 값을 90으로 변경
df
```

```python
df.loc['경석', '체육'] = 110
df
```

```python
df.loc['경석', ['체육', '음악']] = 120, 130
df
```

```python
df1 = df.transpose() # transpose() -> 행열의 위치를 바꿔준다
df1 # 리턴을 받는 값
```

```python
df
```

```python
df2 = df.T # T = transpose()
df2
```

```python
df3 = df.set_index('음악
df3
```

```python
df3 = df.reset_index()
df3
```

```python
df.index = ['옥호', '민자', '영숙'] # 인덱스(행)의 이름을 변경해준다
df
```

```python
df3 = df.reindex(['옥호', '민자', '영숙', '영림', '덕유'], fill_value = 0) # reindex 에러나지 않고 인덱스는 추가된다
df3 # fill_value 는 비어진 값을 채워주는 함수이다
```

```python
df
```

```python
df3 = df.sort_index() # 인덱스 기준으로 오름차순으로 정렬
df3
```

```python
df3 = df.sort_index(ascending = False) # 인덱스 기준으로 내림차순으로 정렬
df3 #ascending -> 오름차순을 뜻함 / False을 주면 반대개념으로 내림차순으로 정렬한다
```

```python
df3 = df.sort_values(by = '수학', ascending = False) # 수학을 기준으로 내림차순으로 정렬
df3 # sort_values 열을 정렬 / (by = '') 열의 무엇을 기준으로 정렬할것인지
```

```python
student1 = pd.Series({'국어':100, '영어':80, '수학':90})
student1
```

```python
percentage = student1 / 100 # 각각의 값에 100으로 나눔
print(percentage)
print(type(percentage))
```

```python
student2 = pd.Series({'영어':80, '국어':100, '수학':90})
student2
```

```python
result = student1 + student2
result
```

```python
student1 - student2
```

```python
student2.drop('영어', inplace=True)
student2
```

```python
result = student1 + student2
result
```

```python
student1['국어'] = float('NaN')
student1
```

```python
result = student1 + student2
print(result)
print(type(result))
```

```python
result = student1.add(student2, fill_value=0) # NaN으로 나타내면 값을 볼수 없기에 fill_value = 0으로 채워 나머지 값을 볼수 있게 해주자
result
```
