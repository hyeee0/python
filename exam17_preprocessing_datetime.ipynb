```python
dates = ['2022-01-01', '2022-03-01', '2022-06-01']
ts_dates = pd.to_datetime(dates) #시각 dtype='datetime64[ns]
ts_dates
```

```python
pr_day = ts_dates.to_period(freq = 'D') #기간 dtype='period[D] 하루
pr_day
```

```python
pr_month = ts_dates.to_period(freq = 'M') #기간 dtype='period[M] 한달
pr_month
```

```python
ts_ms = pd.date_range(start = '2022-01-01', end = None, periods = 6,
                         freq = 'MS', tz = 'Asia/Seoul')
# freq = 'MS' -> month start : 1일 출력/  periods = 6, 6개 자료 -> 6개월
# freq = 'M' -> month : 마지막날로 출력
ts_ms 
```

```python
ts_ms = pd.date_range(start = '2022-01-01', end = None, periods = 6,
                         freq = '3MS', tz = 'Asia/Seoul')
# freq = '3MS' -> month start : 1일 출력, 3개월 간격으로 출력/  periods = 6, 6개 자료 
# freq = 'M' -> month : 마지막날로 출력
ts_ms 
```

```python
ts_ms = pd.date_range(start = '2022-01-01', end = None, periods = 6,
                         freq = '2H', tz = 'Asia/Seoul')
# freq = 'MS' -> month start : 1일 출력/  periods = 6, 6개 자료 -> 6개월
# freq = 'M' -> month : 마지막날로 출력 / freq = '2H' -> 2시간 간격으로 출력
ts_ms 
```
